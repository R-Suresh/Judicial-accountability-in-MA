```python
# import necessary packages 

import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
from collections import Counter
```

```python
appellate = pd.read_json('data/cases.json')
```

```python
appellate
```

```python
# make it easier to analyze / format columns

# checks to make: if len of 1, just pop, otherwise we need to check if a value is a str
for row, _ in appellate.iterrows():
    if len(appellate.loc[row,'case']) > 1:
        for item in appellate.loc[row,'case']:
            if item == '':
                continue 
            elif 'vs' in item:
                appellate.loc[row,'case'] = item
    else:
        appellate.loc[row,'case'] = appellate.loc[row,'case'].pop(-1)
```

```python
appellate
```

```python
# to figure out how many case titles we have to manually input 
for i, val in enumerate(appellate.iloc[:,0].values.tolist()):
    print(f'index:{i} and case title: {val}')
```

```python
# Turn the opinions text into a long string instead of an array of items
for row, _ in appellate.iterrows():
    s = ''
    for item in appellate.loc[row,'text']:
        s+= item
    appellate.loc[row,'text'] = s
```

```python
for row, _ in appellate.iterrows():
    s = ''
    for item in appellate.loc[row,'headnote']:
        s+= item
    appellate.loc[row,'headnote'] = s
```
